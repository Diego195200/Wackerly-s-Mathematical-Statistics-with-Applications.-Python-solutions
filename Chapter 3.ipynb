# Exercise 3.105


```python
# We have two possible distributions for this problem, but first let's look to the graphs
x = range(0,4)

# When Y1: Number of internships who are hired  Y1 ~ Bin(n=3, p=5/8)

plt.bar(x, stats.binom.pmf(x, 3, 5/8))
plt.title('Binomial Distribution')
```

```python
# When Y2: Number of internships who are hired  Y2 ~ HG(N=8, r=3)
plt.bar(x, stats.hypergeom.pmf(x, M=8, n=5, N=3))
plt.title('Hypergeometric Distribution')
```

```python
binomial = stats.binom.pmf([0,1,2,3], 30, 0.2)
print(sum(binomial))

poisson = stats.poisson.pmf([0,1,2,3], 30*0.2)
print(sum(poisson))
```
